# B. 데이터 전처리 – 2021년 이후의 종가 데이터 출력하기

```python
# 패키지 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSV 파일 경로
csv_path = "data/NFLX.csv"

# CSV 파일 읽어오기
df = pd.read_csv(csv_path, usecols=range(5))
```

```python
# 날짜 데이터 변환
df["Date"] = pd.to_datetime(df["Date"])

# 날짜 필터링
df_after_2021 = df[df["Date"] >= "2021-01-01"].copy()

# 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(df_after_2021['Date'], df_after_2021['Close'])


# 그래프 제목 설정
plt.title('Close Prices over Time')

# x축 레이블 설정
plt.xlabel('Date')

# y축 레이블 설정
plt.ylabel('Close')

plt.xticks(rotation=45)

# 그래프 표시
plt.show()
```
